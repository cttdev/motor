# Over Current Protection Calculations
[https://www.st.com/resource/en/application_note/an4999-stspin32f0f0af0b-overcurrent-protection-stmicroelectronics.pdf](https://www.st.com/resource/en/application_note/an4999-stspin32f0f0af0b-overcurrent-protection-stmicroelectronics.pdf)

Constants: \
$I_{max, th} = 60A$ (From architecture calcs) \
$R_{S} = 0.001\Omega$ (From current shunt calculations) \
$\text{OC\_COMP}_{th} = 100mV$ (Smallest threshold value supported by STSPIN) \
$V_{DD} = 3.3v$ (STPSIN Datasheet)

We will choose the triple shunt topology with a bias resistor. Three shunts are already used for current sensing and bias resistor is needed to account for $\frac{1}{3}$ attenuation. (Equation 18)



We are given that,
$$R_B = R_{LP} \cdot (\frac{V_{DD} - \text{OC\_COMP}_{th}}{N_{S} \cdot \text{OC\_COMP}_{th} - I_{max, th} \cdot R_S}) \quad (\text{Table 5})$$

And,
$$R_{LP} \gg R_S$$

Setting $R_{LP} = 2k\Omega$,
$$R_B = 2k\Omega \cdot (\frac{3.3V - 0.1V}{3 \cdot 0.1V - 60A \cdot 0.001\Omega})$$
$$R_B = 26666.6666667\Omega$$
$$R_B = 26.667k\Omega$$

We can round this up to 27k to make the current threshold slightly above 60A.
